```python
timeA <- workouts$workout.aggregates.active_time_total;
timeE <- workouts$workout.aggregates.elapsed_time_total;
dist <- as.double(workouts$workout.aggregates.distance_total);
speed <- as.double(workouts$workout.aggregates.speed_avg);
energy <- as.double(workouts$workout.aggregates.metabolic_energy_total);
ver <- workouts$workout.is_verified;
sel <- !(is.na(timeE)|is.na(timeA)|is.na(energy)|is.na(speed)|is.na(dist)|is.na(ver))
sel1 <- workouts$workout.source=="MapMyRide iPhone";
source <- as.character(workouts$workout.source)
```

```python
mod <- lm(log(speed)~log(energy)+ver,subs=sel&sel1&energy>0&speed>0)
```

```python
summary(mod)
```

```python
table(speed[sel&sel1]>(dist/timeA)[sel&sel1],speed[sel&sel1]>(dist/timeE)[sel&sel1])
```
